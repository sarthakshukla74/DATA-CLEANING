```python
import pandas as pd

df = pd.read_csv('AB_NYC_2019.csv')
df.duplicated().sum()
```

**CHECKING FOR NULL VALUES**

```python
df.isna().sum()
```

```python
df1=df.drop("last_review",axis=1)
df1.isna().sum()
```

**IMPUTATION**

```python
df3=df.copy()
df_mean=df3['reviews_per_month'].mean()
df3['reviews_per_month'].fillna(value=df_mean,inplace=True)
df3.isna().sum()
```

**FOR CATEGORICAL DATA**


*value_counts() makes a list of frequencies of unique values and index[0] gives the most frequent value as the value to be replaced*

```python
df3['last_review'].fillna(df3['last_review'].value_counts().index[0],inplace=True)
df3.isna().sum()
```

**INTERPOLATION**

```python
df4=df.copy()
df4['reviews_per_month'].interpolate(inplace=True)
df4.isna().sum()
```

```python
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
import numpy as np
df=pd.read_csv("scholarship.csv")
display(df)
```

```python
plt.subplot(1,2,2)
sns.distplot(df["semester_percentage"])
plt.subplot(1,2,1)
sns.distplot(df["scholarship_exam_marks"])
plt.show()
```

```python
df.describe()
```

```python
highest_allowed=df["semester_percentage"].mean()+3*df["semester_percentage"].std()
lowest_allowed=df["semester_percentage"].mean()-3*df["semester_percentage"].std()
```

```python
outliers=df[(df["semester_percentage"]<highest_allowed) | (df["semester_percentage"]>lowest_allowed) ]
print(outliers)
```

**TRIMMING**

```python
trimmed_data=df[(df["semester_percentage"]<highest_allowed) & (df["semester_percentage"]>lowest_allowed) ]
print(trimmed_data)
```

**DATA CLEANING USING Z-TEST**

```python
df["z-score"]=(df["semester_percentage"]-df["semester_percentage"].mean())/df["semester_percentage"].std()
outliers_zscore=df[(df["z-score"]>-3) & (df["z-score"]<3)]
print(outliers_zscore)
```

```python
import numpy as np
import pandas as pd
import seaborn as sns
from matplotlib import pyplot as plt

data=pd.read_csv('scholarship.csv')
```

```python
data['scholarship_exam_marks'].skew()
```

```python
data['scholarship_exam_marks'].describe()
```

```python
sns.boxplot(data['scholarship_exam_marks'])
```

**CALCULATING IQR**

```python
Q1=data['scholarship_exam_marks'].quantile(0.25)
Q3=data['scholarship_exam_marks'].quantile(0.75)
IQR=Q3-Q1
lb=Q1-1.5*IQR
ub=Q3+1.5*IQR
outliers=data[(data['scholarship_exam_marks']<lb) | (data['scholarship_exam_marks']>ub)]
print(outliers)
```

**DATA VISUALIZATION**

```python
new_data=data[(data['scholarship_exam_marks']>lb) & (data['scholarship_exam_marks']<ub)]
plt.figure(figsize=(16,8))
plt.subplot(2,2,1)
sns.distplot(data['scholarship_exam_marks'])
plt.subplot(2,2,2)
sns.boxplot(data['scholarship_exam_marks'])
plt.subplot(2,2,3)
sns.distplot(new_data['scholarship_exam_marks'])
plt.subplot(2,2,4)
sns.boxplot(new_data['scholarship_exam_marks'])
plt.show()
```

**USING PERCENTILE**

```python
max_limit=data['scholarship_exam_marks'].quantile(0.99)
min_limit=data['scholarship_exam_marks'].quantile(0.02)
up_data=data[(data['scholarship_exam_marks']>=min_limit) & (data['scholarship_exam_marks']<=max_limit)]
print(f"Original data shape{data.shape}")
print(f"Update data shape{up_data.shape}")
```

```python
sns.boxplot(x=up_data['scholarship_exam_marks'])
plt.show()
```
